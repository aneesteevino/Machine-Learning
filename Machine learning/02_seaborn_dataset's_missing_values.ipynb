```python
# Visual Inspection of  missing values

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load titanic dataset
planets = sns.load_dataset('planets')

# Visualize the data
plt.figure(figsize=(8, 5))
sns.heatmap(planets.isnull(), cbar=False)
plt.show()
```

```python
planets.isnull().sum().sort_values(ascending=False)
```

```python
# filling values by mean


mean = planets['distance'].mean()
planets['distance'] = planets['distance'].fillna(mean)
planets['mass'] = planets['mass'].fillna(mean)
planets['orbital_period'] = planets['orbital_period'].fillna(mean)
planets.isnull().sum().sort_values(ascending=False)
```
