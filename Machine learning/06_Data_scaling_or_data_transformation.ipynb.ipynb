# 1. Data Standardization
`Data Standardization`  ek technique hai jisse hum data ko ek standard scale pe la sakte hain. or 
data ko standard scale pe laane ka matlab hai ki data ka mean 0 or standard deviation 1 ho jaye.
Data standardization ka formula hai:
$$ z = (x- \mu) / \sigma $$

```python
# import libraries

import pandas as pd
from sklearn.preprocessing import StandardScaler


data ={
    "age":[10,15,20,25,30,35,40,45,50,55],
   "height":[70,75,80,85,90,95,100,105,110,115],
   "weight":[30,35,40,45,50,55,60,65,70,75]

}


df=pd.DataFrame(data)

scaler=StandardScaler()

standardizing_data=scaler.fit_transform(df)
standardized_data=pd.DataFrame(standardizing_data,columns=["age","height","weight"])

standardized_data





```

# Min-Max Scaler

`Min-Max Scaler` ek technique hai jisse hum data ko 0 se 1 ke beech me scale kar sakte hain.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler,MaxAbsScaler

data={
    "Age":[1,2,3,4],
    "height":[5,6,7,8]
}

df=pd.DataFrame(data)

MinScaler=MinMaxScaler()
Minmax_scaling=MinScaler.fit_transform(df)
Minmax_scaled_data=pd.DataFrame(Minmax_scaling,columns=["Age","height"])

Minmax_scaled_data.head()
```

# Max Absolute Scaler

`Max Absolute Scaler` ek technique hai jisse hum data ko -1 se 1 ke beech me scale kar sakte hain.


```python
import pandas as pd
from sklearn.preprocessing import MaxAbsScaler

data={
    "Age":[12,13,15],
    "height":[5,4,7]

}

df=pd.DataFrame(data)

Scaler=MaxAbsScaler()

MaxAbsScaling=Scaler.fit_transform(df)

MaxAbsScaled_data=pd.DataFrame(MaxAbsScaling,columns=["Age","height"])
MaxAbsScaled_data
```
